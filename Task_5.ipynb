```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr= dataset.corr()
corr
sns.heatmap(corr, annot= True)
```

```python
plt.hist(dataset['Discount'], bins=10)
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset.groupby(dataset['Region'])['Profit'].mean().plot.bar()
```

```python
fig = plt.figure(figsize=(50,10))
plt.pie(dataset['Region'].value_counts(), labels=dataset['Region'].value_counts().index,autopct='%1.1f%%')
plt.title('Percentage of deals in different Regions in USA')
plt.show()
```

```python
dataset['Sub-Category'].value_counts().plot(kind='bar')
```

```python
dataset.groupby(dataset['Sub-Category'])["Sales"].max().plot.bar()
```

```python
fig = plt.figure(figsize = (50,20))
plt.pie(dataset['Sub-Category'].value_counts(), labels = dataset['Sub-Category'].value_counts().index, autopct='%1.1f%%')
plt.title('Percentage of Deals in different Sub-Categories in USA')
plt.show()
```

```python
dataset['State'].value_counts()
```

```python
dataset['State'].value_counts().plot(kind='bar', figsize=(20,10), color='Red')
```

```python
fig = plt.figure(figsize = (60,25))
plt.pie(dataset['City'].value_counts(),labels=dataset['City'].value_counts().index,autopct='%1.1f%%')
plt.title("Percentage of Deals in different cities in USA")
plt.show()
```

```python
dataset.hist(figsize=(12,12),bins=50)
plt.show()
```

```python
pd.DataFrame(dataset.groupby('State').sum())['Profit'].sort_values(ascending=True)
```

```python
pd.DataFrame(dataset.groupby('State').sum())['Discount'].sort_values(ascending=True)
```

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = dataset, color = 'b', label= 'Discount')
plt.legend()
```
